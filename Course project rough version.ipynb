Neurotism is one of these big 5 measures, and it tracks how likely someone is to feel negative emotions. So, high neurotism could be something like a curse, a disposition to be sad and stressed all throught life. Studies have shown though, that change is possible, but mostly from changes in situation, which isn't exactly helpful to hear.
```python
import pandas as pd

cols = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/var_names.csv")
data = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv",
                   na_values=["9999", "", " ", "Presented but no response", "NA"],low_memory=False)
empty = (data.isna().sum()==data.shape[0])
df = data[empty.index[~empty]] 
df = df[df.REMOVE_case == 'No'].copy()
df = df[df.SURVEY_collection_year == 2023].copy()

missing_limit = 999
keep = df.isna().sum() < missing_limit
keep = keep.index[keep] #I have literally no idea what this line does, I just copied it from the project intro
df = df[keep].copy()
df.shape
```

```python
import statsmodels.formula.api as smf

focus = 'PSYCH_big_five_inventory_neurotic_score'
formula = f'{focus} ~ '

for name in df.columns.values:
    if 'HEALTH_hampson' in str(name):
        formula += f' + C({name}, Treatment(reference="Strongly agree"))'

res = smf.ols(formula=formula, data=df).fit().summary2()
filtered = res.tables[1]
filtered = filtered[filtered['P>|t|'] <= 0.1]
filtered
```

```python
import statsmodels.formula.api as smf

focus = 'PSYCH_big_five_inventory_neurotic_score'
formula = f'{focus} ~ '


for name in df.columns.values:
    if 'HEALTH_hampson' in str(name):
        formula += f' + C({name}, Treatment(reference="Strongly agree"))'
        
for name in df.columns.values:
    if 'CONNECTION' in str(name):
        formula += f' + {name}'


res = smf.ols(formula=formula, data=df).fit().summary2()
filtered = res.tables[1]
filtered = filtered[filtered['P>|t|'] <= 0.01]
filtered
```

```python
import statsmodels.formula.api as smf

focus = 'PSYCH_big_five_inventory_neurotic_score'
formula = f'{focus} ~ C(HEALTH_hampson_good_health_practices_scale_diet) + C(HEALTH_hampson_good_health_practices_scale_sleep) + C(HEALTH_hampson_good_health_practices_scale_limit_foods)'

res = smf.ols(formula=formula, data=df).fit().summary2()
filtered = res.tables[1]
filtered = filtered[filtered['P>|t|'] <= 0.1]
filtered
```
