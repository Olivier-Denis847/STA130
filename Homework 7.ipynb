- Difference Between Simple and Multiple Linear Regression:
    Simple Linear Regression: Involves one independent variable and one dependent variable. The model is ( y = b_0 + b_1x ).
    Multiple Linear Regression: Involves one dependent variable and multiple independent variables. The model is ( y = b_0 + b_1x_1 + b_2x_2 + \ldots + b_nx_n ).
- Continuous vs. Indicator Variables in Simple Linear Regression:
    Continuous Variables: Numerical values that can take any value within a range. The coefficient represents the change in the dependent variable for a one-unit change in the predictor.
    Indicator Variables: Binary variables (0 or 1) representing categorical data. The coefficient represents the difference in the dependent variable between the two categories.
- Multiple Linear Regression with Continuous and Indicator Variables:
    Model: Can include both types of variables. The coefficients for continuous variables represent the change in the dependent variable per unit change, while coefficients for indicator variables represent the difference between categories.
    Interaction Terms: Adding an interaction term between a continuous and an indicator variable allows the model to capture how the effect of the continuous variable changes across categories.
- Interaction Between Continuous and Indicator Variables:
    Definition: An interaction term is the product of a continuous variable and an indicator variable.
    Interpretation: The interaction coefficient represents how the effect of the continuous variable on the dependent variable changes depending on the category of the indicator variable.
- Simple Logistic Regression Model:
    Purpose: Used for binary classification problems.
    Example: Predicting whether a student will pass or fail an exam based on hours studied.
    Model: ( \text{logit}(p) = b_0 + b_1x ), where ( p ) is the probability of passing.
- Continuous Variables in Logistic Regression:
    Inclusion: Logistic regression can include continuous variables as predictors.
    Interpretation: The coefficients represent the change in the log-odds of the outcome for a one-unit increase in the predictor.
- Log Odds:
    Definition: The natural logarithm (base ( e )) of the odds of an event occurring.
    Example: A log odds of -4 corresponds to a probability of approximately 1.8%, not ( 10^{-4} ).

```python
#AI code is unresonably good, I have no future
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split

# Function to simulate model effectiveness with different random seeds
def simulate_models(random_seeds):
    results = []

    for seed in random_seeds:
        np.random.seed(seed)
        
        # Assuming pokeaman is a DataFrame, let's create a sample DataFrame for demonstration
        data = {
            'HP': np.random.randint(50, 150, size=100),
            'Attack': np.random.randint(50, 150, size=100),
            'Defense': np.random.randint(50, 150, size=100),
            'Speed': np.random.randint(50, 150, size=100),
            'Legendary': np.random.choice([0, 1], size=100),
            'Sp. Def': np.random.randint(50, 150, size=100),
            'Sp. Atk': np.random.randint(50, 150, size=100),
            'Type 1': np.random.choice(['Normal', 'Fire', 'Water'], size=100),
            'Type 2': np.random.choice(['None', 'Fire', 'Water'], size=100),
            'Generation': np.random.randint(1, 6, size=100)
        }

        pokeaman = pd.DataFrame(data)

        # Data Preparation
        fifty_fifty_split_size = int(pokeaman.shape[0] * 0.5)
        pokeaman.fillna('None', inplace=True)
        pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=fifty_fifty_split_size)

        # Model 5 Specification and Fitting
        model5_linear_form = 'HP ~ Attack + Defense + Speed + Legendary'
        model5_linear_form += ' + Q("Sp. Def") + Q("Sp. Atk")'
        model5_linear_form += ' + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

        model5_spec = smf.ols(formula=model5_linear_form, data=pokeaman_train)
        model5_fit = model5_spec.fit()
        yhat_model5 = model5_fit.predict(pokeaman_test)
        y = pokeaman_test.HP
        in_sample_r2_model5 = model5_fit.rsquared
        out_of_sample_r2_model5 = np.corrcoef(y, yhat_model5)[0, 1]**2

        # Model 6 Specification and Fitting
        model6_linear_form = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk")'
        model6_linear_form += ' + I(Q("Type 1")=="Normal")'
        model6_linear_form += ' + I(Q("Type 1")=="Water")'
        model6_linear_form += ' + I(Generation==2)'
        model6_linear_form += ' + I(Generation==5)'

        model6_spec = smf.ols(formula=model6_linear_form, data=pokeaman_train)
        model6_fit = model6_spec.fit()
        yhat_model6 = model6_fit.predict(pokeaman_test)
        in_sample_r2_model6 = model6_fit.rsquared
        out_of_sample_r2_model6 = np.corrcoef(y, yhat_model6)[0, 1]**2

        # Model 7 Specification and Fitting
        model7_linear_form = 'HP ~ Attack * Speed * Q("Sp. Def") * Q("Sp. Atk")'
        model7_linear_form += ' + I(Q("Type 1")=="Normal")'
        model7_linear_form += ' + I(Q("Type 1")=="Water")'
        model7_linear_form += ' + I(Generation==2)'
        model7_linear_form += ' + I(Generation==5)'

        model7_spec = smf.ols(formula=model7_linear_form, data=pokeaman_train)
        model7_fit = model7_spec.fit()
        yhat_model7 = model7_fit.predict(pokeaman_test)
        in_sample_r2_model7 = model7_fit.rsquared
        out_of_sample_r2_model7 = np.corrcoef(y, yhat_model7)[0, 1]**2

        # Model 7 with Centering and Scaling Specification and Fitting
        model7_linear_form_CS = 'HP ~ scale(center(Attack)) * scale(center(Speed))'
        model7_linear_form_CS += ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
        model7_linear_form_CS += ' + I(Q("Type 1")=="Normal")'
        model7_linear_form_CS += ' + I(Q("Type 1")=="Water")'
        model7_linear_form_CS += ' + I(Generation==2)'
        model7_linear_form_CS += ' + I(Generation==5)'

        model7_CS_spec = smf.ols(formula=model7_linear_form_CS, data=pokeaman_train)
        model7_CS_fit = model7_CS_spec.fit()
        yhat_model7_CS = model7_CS_fit.predict(pokeaman_test)
        in_sample_r2_model7_CS = model7_CS_fit.rsquared
        out_of_sample_r2_model7_CS = np.corrcoef(y, yhat_model7_CS)[0, 1]**2

        results.append({
            "Seed": seed,
            "Model": "Model 5",
            "In Sample R-squared": in_sample_r2_model5,
            "Out of Sample R-squared": out_of_sample_r2_model5
        })
        
        results.append({
            "Seed": seed,
            "Model": "Model 6",
            "In Sample R-squared": in_sample_r2_model6,
            "Out of Sample R-squared": out_of_sample_r2_model6
        })
        
        results.append({
            "Seed": seed,
            "Model": "Model 7",
            "In Sample R-squared": in_sample_r2_model7,
            "Out of Sample R-squared": out_of_sample_r2_model7
        })
        
        results.append({
            "Seed": seed,
            "Model": "Model 7 (Centered & Scaled)",
            "In Sample R-squared": in_sample_r2_model7_CS,
            "Out of Sample R-squared": out_of_sample_r2_model7_CS
        })

    return pd.DataFrame(results)

# Simulate models with different random seeds
random_seeds = [42, 100, 200]
simulation_results = simulate_models(random_seeds)

print(simulation_results)
```
